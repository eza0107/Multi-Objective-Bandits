```python
# Pareto Front of SGT
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(12, 6), dpi=200)
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 15,
        }

for i in range(2):
    plt.subplot(1,2,i+1)
    if i==0:
        # non-conflict version
        x = np.array([0,0,-50,-50])
        y = np.array([0,0,50, 50])
        x = (x - np.max(x))/(np.max(x) - np.min(x))
        plt.plot(x,y, color = 'red')
        plt.plot([x[0]],[y[0]], color = '#1f77b4', marker = "o", markersize = 5, markerfacecolor = '#1f77b4')
        plt.plot([x[2]],[y[2]], color = '#1f77b4', marker = "o", markersize = 5, markerfacecolor = '#1f77b4')
        plt.annotate("Decks A,B", xy = (-0.25,0))
        plt.annotate("Decks C,D", xy = (-0.95,50))
        plt.title("Conflict version " , fontdict = font)
    else:
        x = [-50,-50,0,0]
        y = [0,0,50,50]
        x = (x - np.max(x))/(np.max(x) - np.min(x))
        plt.plot(x,y, color = 'black')
        plt.plot([x[0]],[y[0]], color = '#1f77b4', marker = "o", markersize = 5, markerfacecolor = '#1f77b4')
        plt.plot([x[2]],[y[2]], color = 'red', marker = "o", markersize = 5, markerfacecolor = "red")
        plt.annotate("Decks A,B", xy = (-0.95,0))
        plt.annotate("Decks C,D", xy = (-0.25,50))
        plt.title("Non-conflict version " , fontdict = font)
```
