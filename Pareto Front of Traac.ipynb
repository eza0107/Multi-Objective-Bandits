```python
# plotting Pareto Front for Weather Task
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(20, 14), dpi=200)

x = np.array([[-0.2, -0.5, -0.8],
     [-0.2, -0.5, -0.8],
     [-0.8, -0.5, -0.2],
     [-0.8, -0.5, -0.2]])
for row in x:
    x = (x - np.max(x))/(np.max(x)-np.min(x))
print(x)   
y = np.array([[0.8, 0.5, 0.2],
     [0.2, 0.5, 0.8],
     [0.2, 0.5, 0.8],
     [0.8, 0.5, 0.2]])


font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 25,
        }
for i in range(4):
    plt.subplot(2,2,i+1)
    if i%2==1:
        plt.plot(x[i],y[i],"-", color = 'red', marker = ",", markersize = 5, markerfacecolor = "blue")
        plt.title("Conflict Block " + str(int((i+1)/2)), fontdict = font)
    else:
        plt.plot(x[i],y[i],"-", color = 'black', marker = "o", markersize = 10, markerfacecolor = '#1f77b4')
        plt.title("Congruent Block " + str(int((i+2)/2)), fontdict = font)
    if i==0 or i==2:
        plt.plot(x[i][i], y[i][i], color = 'red', marker = "o", markersize = 10, markerfacecolor = 'red')
    else:
        for j in range(3):
            plt.plot(x[i][j], y[i][j], color = '#1f77b4', marker = "o", markersize = 10, markerfacecolor = '#1f77b4')
    if i%2==0:
        plt.ylabel("Objective 2", fontsize = 25)
    if i>=2:
        plt.xlabel("Objective 1", fontsize = 25)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
```
