```python
# plotting Pareto Front for Weather Task
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(20, 14), dpi=200)

x = [[0.0, 0.3, 0.6],
     [0.0, 0.3, 0.6],
     [0.6, 0.3, 0.0],
     [0.6, 0.3, 0.0]]
y = [[0.0, 1.65, 3.3],
     [3.3, 1.65, 0.0],
     [3.3, 1.65, 0.0],
     [0.0, 1.65, 3.3]]

font= {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 25,
        }
for i in range(4):
    plt.subplot(2,2,i+1)
    if i%2==1:
        plt.plot(x[i],y[i],"-", color = 'red', marker = "o", markersize = 5, markerfacecolor = "black")
        plt.title("Conflict Block " + str(int((i+1)/2)), fontdict = font)
    else:
        plt.plot(x[i],y[i],"s", color = 'black', marker = "o", markersize = 5, markerfacecolor = "black")
        plt.title("Congruent Block " + str(int((i+2)/2)), fontdict = font)
    if i==0:
        plt.plot(x[i][0], y[i][0], color = 'red', marker = "o", markersize = 10, markerfacecolor = 'red')
    if i==2:
        plt.plot(x[i][2], y[i][2], color = 'red', marker = "o", markersize = 10, markerfacecolor = 'red')
    if i%2==0:
        plt.ylabel("Reward regret", fontsize = 25)
    if i>=2:
        plt.xlabel("Threat regret", fontsize = 25)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
plt.suptitle("Pareto Fronts of trAAC Task", fontsize  = 35)
```
