```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 5), dpi=200)
plt.figure(1)

y = np.array([0, 4, 9,10,6])
x = np.array([0,-1, -4,-10,-3])

#scaling 
x=x
y = y
#plt.plot([x[1], x[2],x[3]], [y[1], y[2], y[3]], '-', color = 'red')

plt.plot([x[0], x[1], x[2],x[3]], [y[0], y[1], y[2], y[3]], '-', color = 'red')


plt.plot([x[0], x[3]], [y[0] ,y[3]], '-', color = 'black')
plt.legend(['Pareto Front', '$\Delta$'])
y1 = y[:-1]
y2 = [y[0]]
m = (y[3] - y[0])/(x[3] - x[0])
b = (y[0]*x[3] - y[3]*x[0])/(x[3] - x[0])

y2.append(m*x[1]+b)
y2.append(m*x[2]+b)
y2.append(y[3])
print(np.shape(y1), np.shape(y2))

plt.fill_between(x[:-1],y1,y2, color = 'lightgrey')
plt.scatter(x,y)

annotations=["1","2","3","4", "5"]
for i, label in enumerate(annotations):
    plt.annotate(label, (x[i]+0.2, y[i]))
plt.xlabel("Objective 1")
plt.ylabel("Objective 2")
plt.show()

```
