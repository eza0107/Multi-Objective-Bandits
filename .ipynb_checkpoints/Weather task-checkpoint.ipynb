```python
import numpy as np
import math
from matplotlib import pyplot as plt
from random import choices
S = 10

T = 100 # for the long term make T larger. 
#---------------------parameters for epsilon - greedy-----------------------
alpha = 0.1
eps = 0.3
gamma = 0.0
tau = 1
#---------------------------------------------------------------------------
rewards = np.array([0,2,4,6])
threats = np.array([1,-1])
max_reward_mean = rewards*0.9
max_threat_mean = threats[0]*0.9
#---------------------------------------------------------------------------
def linear_scalar (Q, w, epsilon):
    temp = Q[0,:]*w[0] + Q[1,:]*w[1]
    if np.random.uniform(0,1) < epsilon:
        action = np.random.randint(0,np.size(Q,1))
    else:
        action = np.random.choice(np.array(np.where(temp == np.amax(temp))).flatten())
    return action

def softmax_one (Q, w, tau):
    temp = (Q[0,:]*w[0] + Q[1,:]*w[1])/tau 
    ex = np.exp(temp - np.max(temp))
    weights = ex/np.sum(ex)
    #print(weights)
    population = np.arange(np.size(Q,1))
    
    action = choices(population, weights)
    return action[0]

def softmax_two (Q, w, tau):
    ex_zero =  np.exp((Q[0,:] - np.max(Q[0,:]))/tau)
    weights_zero = ex_zero/np.sum(ex_zero)
    ex_one =  np.exp((Q[1,:] - np.max(Q[1,:]))/tau)
    weights_one = ex_one/np.sum(ex_zero)
    
    weights = w[0]*weights_zero + w[1]*weights_one
    #print('-----------')
    #print(w)
    #print(weights_zero, weights_one)
    #print(weights)
    population = np.arange(np.size(Q,1))

    action = choices(population, weights)
    
    return action[0]
# put the choosing of maximum function inside here.     
def update(curr_action, Q, next_action, r, obj,alpha):
    #predict = Q[obj, curr_action]
    target = r+ gamma*Q[obj, next_action]
    Q[obj, curr_action] = Q[obj, curr_action]*(1-alpha) + alpha*target

def step(curr_action, stage):
    p_good = 0.1*(curr_action+1)
    if np.random.uniform(0,1) < p_good: # good outcome
        threat = 1
        reward = 0
    else: # bad outcome
        threat = -1
        reward = stage*2
    return reward, threat
#----------------TRAINING STARTS HERE --------------------
# partition parameter
p = 0.02
W = int(1/p) + 1

Q = np.zeros((S,2,9,W,4))

reward_regret = np.zeros((S,T,W,4))
threat_regret = np.zeros((S,T,W,4))



for stage in range(4):
    for w in range(W):
        if w%10==0:
            print(w)
        weight = [w*p, 1.0-w*p]
        #print(weight)
        for s in range(S):
            for t in range(T):
                curr_action = softmax_one(Q[s,:,:,w,stage], weight, tau)
                #print("action = ", curr_action)
                #actions[s,curr_action,w]+=1
                reward, threat = step(curr_action, stage)

                #print("reward = ", np.round(reward,3))
                #print("threat = ", np.round(threat,3))
                # not the weighted Q here
                next_action = linear_scalar(Q[s,:,:,w,stage], weight, 0.0)

                update(curr_action, Q[s,:,:,w], next_action, reward, 0,alpha)
                update(curr_action, Q[s,:,:,w], next_action, threat, 1,alpha)
                #print(np.round(Q[s,:,:,w],3))
                reward_regret[s,t,w,stage] = max_reward_mean[stage] - reward
                threat_regret[s,t,w,stage] = max_threat_mean - threat
```

```python
rew_regret = np.zeros((9,4))
thr_regret = np.zeros((9,4))

for i in range(9):
    for s in range(4):
        rew_regret[i,s] = 0.2*s*i
        thr_regret[i,s] = 1.6 - 0.2*i
```

```python
# plotting asymptotics against the weight.
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(16, 10), dpi=80)
plt.figure(1)

# Pareto front plotter
# returns true only if (a,b) is not Pareto dominated by (a1,b1)
def isPareto(a,b,a1,b1):
    if a1 > a:
        return True
    elif b1 > b:
        return True
    else:
        return False
#print(isPareto(0.2,0.2,0.2,0.2))


lbls = ['1 ', '2', '3', '4', '5', '6', '7', '8', '9']
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 25,
        }
# for i in range(9):
#     plt.text(thr_regret[i], rew_regret[i], lbls[i], fontdict = font)


x = np.arange(T)
cum_rew_regret = np.zeros((T,W,4))
cum_thr_regret = np.zeros((T,W,4))
print(np.shape(reward_regret))

for stage in range(4):
    for w in range(W):
        cum_rew_regret[:,w,stage] = np.cumsum(np.mean(reward_regret[:,:,w,stage], 0))
        cum_thr_regret[:,w,stage] = np.cumsum(np.mean(threat_regret[:,:,w,stage], 0))

for t in range(T):
    cum_rew_regret[t,:,:]/=(t+1)
    cum_thr_regret[t,:,:]/=(t+1)

for stage in range(4):
    plt.subplot(2,2,stage+1)
    y = cum_rew_regret[T-1,:, stage] # make T larger average over last few time steps
    x = cum_thr_regret[T-1,:, stage]
    annotations=[]
    n = np.size(x)
    for i in range(n):
        count = 0
        for j in range(n):
            if  isPareto(x[i], y[i], x[j], y[j]):
                count+=1
        if count==n-1:
            
            plt.plot(x[i], y[i],marker="o",markersize=10, markeredgecolor="red",
                markerfacecolor="orange")

    plt.plot(x[0], y[0], marker = "o", markersize = 20, markerfacecolor = "blue")
    plt.text(x[-1], y[-1], "reward = " + str(2*stage))
    plt.plot(x[n-1], y[n-1], marker = "o", markersize = 20, markerfacecolor = "green")

    print(thr_regret[:,stage], rew_regret[:,stage])
    plt.plot(thr_regret[:,stage],rew_regret[:,stage], "kx", marker = "o", markersize = 5, markerfacecolor = "black")
    for w in range(W):
        f = format(w*p, ".2f")
        annotations = np.append(annotations,"w = "+str(f) )
    plt.scatter(x,y, s=40)


    plt.xlabel("Threat regret")
    plt.ylabel("Reward regret")
plt.suptitle("Pareto front of Threat-Reward Regret Pairs: Softmax with tau = " +str(eps),fontsize=15)
plt.show()
```
